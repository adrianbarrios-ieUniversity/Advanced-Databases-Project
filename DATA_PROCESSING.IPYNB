```python
import pandas as pd
import xlrd
import os

data_dir = 'DATASETS'

files = [
    'CONGENITAL_SYPHILIS_CASES_REPORTED.xlsx',
    'FERTILITY_RATE_BIRTHS_PER_WOMAN.xls',
    'GOVERNMENT_HEALTH_EXPENDITURE_PER_CAPITA.xlsx',
    'HOSPITAL_BEDS_PER_10000_POPULATION.xlsx',
    'INFANT_MORTALITY_RATE_PER_1000_BIRTHS.xlsx',
    'LIFE_EXPECTANCY_AT_BIRTH.xlsx',
    'MALARIA_INCIDENCE_PER_1000_POPULATION_AT_RISK.xlsx',
    'OVERWEIGHT_PERCENTAGE_AMONG_ADULTS_AGE_STANDARDIZED.xlsx',
    'TUBERCULOSIS_INCIDENT_CASES.xlsx'
]

datasets = {}
for file in files:
    file_path = os.path.join(data_dir, file)
    if file.endswith('.xls') or file.endswith('.xlsx'):
        datasets[file] = pd.read_excel(file_path) # This line was asking for xlrd.
```
