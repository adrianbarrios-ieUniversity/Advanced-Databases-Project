```python
import pandas as pd
```

```python
import os

data_dir = '../CLEANED_DATASETS/'

files = [
    'life_expectancy_clean.xlsx',
]

datasets = {}
for file in files:
    file_path = os.path.join(data_dir, file)
    if file.endswith('.xls') or file.endswith('.xlsx'):
        datasets[file] = pd.read_excel(file_path) # This line was asking for xlrd.
```

```python
LIFE_EXPECTANCY =datasets.get("life_expectancy_clean.xlsx")
LIFE_EXPECTANCY.head()
```

```python
# Rename columns
df = LIFE_EXPECTANCY.rename(columns={
    "CountryName": "country",
    "Year": "year",
    "IndicatorName": "indicator",
    "indicatorValue": "indicator_value"
})


# Reorder to match staging table
df = df[[
    "country",
    "year",
    "indicator",
    "indicator_value"
]]
```

```python
df.to_excel("../CLEANED_DATASETS/life_expectancy_clean.xlsx", 
           index=False, 
           engine="openpyxl")
```

```python
# Check for matching country and year combinations
matching_combinations = df.groupby(['country', 'year']).size()
print(f"Total unique country-year combinations: {len(matching_combinations)}")
print("\nFirst few combinations:")
print(matching_combinations.head(10))

# Check for duplicates
duplicates = matching_combinations[matching_combinations > 1]
print(f"\nDuplicate country-year combinations: {len(duplicates)}")
if len(duplicates) > 0:
    print(duplicates)
```
