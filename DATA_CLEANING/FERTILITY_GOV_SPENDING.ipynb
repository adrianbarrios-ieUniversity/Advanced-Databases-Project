```python
import pandas as pd
```

```python
# Load the fertility rate dataset
df = pd.read_excel('../DATASETS/FERTILITY_RATE_BIRTHS_PER_WOMAN.xls')
df
```

```python
# Inspect column names
print("Column names:")
print(df.columns.tolist())
```

```python
# Count NAs
print("NA count per column:")
print(df.isna().sum())
```

```python
# Reshape the dataset from wide to long format
# Keep country info columns and melt year columns into a single column
df_long = pd.melt(
    df,
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='Year',
    value_name='Fertility_Rate'
)

# Convert Year to integer
df_long['Year'] = df_long['Year'].astype(int)

df_long
```

```python
# Load the government health expenditure dataset
df_gov = pd.read_excel('../DATASETS/GOVERNMENT_HEALTH_EXPENDITURE_PER_CAPITA.xlsx')
df_gov
```

```python
# Inspect column names
print("Column names:")
print(df_gov.columns.tolist())
```

```python
# Count NAs
print("NA count per column:")
print(df_gov.isna().sum())
```

```python
# Save both datasets to the CLEANED_DATASETS folder
df_long.to_csv('../CLEANED_DATASETS/fertility_rate_cleaned.csv', index=False)
df_gov.to_csv('../CLEANED_DATASETS/government_health_expenditure_cleaned.csv', index=False)

print("Datasets saved successfully!")
print(f"Fertility Rate: {len(df_long)} rows")
print(f"Government Health Expenditure: {len(df_gov)} rows")
```

```python
# Rename columns in fertility dataset
df_long_renamed = df_long.rename(columns={
    'Country Name': 'CountryName',
    'Country Code': 'CountryCode',
    'Indicator Name': 'IndicatorName',
    'Fertility_Rate': 'IndicatorValue'
})

# Drop Indicator Code column
df_long_renamed = df_long_renamed.drop(columns=['Indicator Code'])

print("Fertility dataset columns after renaming:")
print(df_long_renamed.columns.tolist())
df_long_renamed.head()
```

```python
# Rename columns in government spending dataset
df_gov_renamed = df_gov.rename(columns={
    'Indicator': 'IndicatorName',
    'ParentLocation': 'Region',
    'SpatialDimValueCode': 'CountryCode',
    'Location': 'Country',
    'Period': 'Year',
    'FactValueNumeric': 'IndicatorValue'
})

# Drop IndicatorCode column
df_gov_renamed = df_gov_renamed.drop(columns=['IndicatorCode'])

print("Government spending dataset columns after renaming:")
print(df_gov_renamed.columns.tolist())
df_gov_renamed.head()
```

```python
# Save renamed datasets to the CLEANED_DATASETS folder
df_long_renamed.to_csv('../CLEANED_DATASETS/fertility_rate_cleaned.csv', index=False)
df_gov_renamed.to_csv('../CLEANED_DATASETS/government_health_expenditure_cleaned.csv', index=False)

print("Renamed datasets saved successfully!")
print(f"Fertility Rate columns: {df_long_renamed.columns.tolist()}")
print(f"Government Health Expenditure columns: {df_gov_renamed.columns.tolist()}")
```
