```python
import pandas as pd
import os

data_dir = '../DATASETS'

files = [
    'CONGENITAL_SYPHILIS_CASES_REPORTED.xlsx',
]

datasets = {}
for file in files:
    file_path = os.path.join(data_dir, file)
    if file.endswith('.xls') or file.endswith('.xlsx'):
        datasets[file] = pd.read_excel(file_path) # This line was asking for xlrd.
```

```python
CONGENITAL_SYPHILIS_RAW =datasets.get("CONGENITAL_SYPHILIS_CASES_REPORTED.xlsx")
CONGENITAL_SYPHILIS_RAW.head()
CONGENITAL_SYPHILIS_RAW.columns = CONGENITAL_SYPHILIS_RAW.iloc[1]

# Remove the first 3 rows (0, 1, 2)
CONGENITAL_SYPHILIS_RAW = CONGENITAL_SYPHILIS_RAW.iloc[2:].reset_index(drop=True)
```

```python
cols_to_drop = [
    "Dim1 type",
    "Dim1",
    "Dim1ValueCode",
    "Dim2 type",
    "Dim2",
    "Dim2ValueCode",
    "Dim3 type",
    "Dim3",
    "Dim3ValueCode",
    "FactValueUoM",
    "FactValueNumericLowPrefix",
    "FactValueNumericLow",
    "FactValueNumericHighPrefix",
    "FactValueNumericHigh"
]
```

```python
CONGENITAL_SYPHILIS = CONGENITAL_SYPHILIS_RAW.drop(columns=cols_to_drop)
```

```python
CONGENITAL_SYPHILIS.head()
```

AI Data Transformation

```python
df = CONGENITAL_SYPHILIS.copy()

# -------------------------------------------------
# 1️⃣ Clean column names
# -------------------------------------------------
df.columns = df.columns.str.strip()

# -------------------------------------------------
# 2️⃣ Keep only relevant columns for the ER model
# -------------------------------------------------
keep_cols = [
    # Region
    "ParentLocationCode",
    "ParentLocation",

    # Country
    "SpatialDimValueCode",
    "Location",

    # Date
    "Period",
    "Period type",

    # Indicator
    "IndicatorCode",
    "Indicator",

    # Fact value
    "FactValueNumeric",
    "Value"
]

df = df[[c for c in keep_cols if c in df.columns]].copy()

# -------------------------------------------------
# 3️⃣ Keep only Year data
# -------------------------------------------------
if "Period type" in df.columns:
    df = df[df["Period type"] == "Year"].copy()
    df.drop(columns=["Period type"], inplace=True)

# -------------------------------------------------
# 4️⃣ Choose correct numeric value
# -------------------------------------------------
if "FactValueNumeric" in df.columns:
    df["Value"] = df["FactValueNumeric"].combine_first(df.get("Value"))
    df.drop(columns=["FactValueNumeric"], inplace=True)

# -------------------------------------------------
# 5️⃣ Rename columns to match ER model
# -------------------------------------------------
df.rename(columns={
    "ParentLocationCode": "RegionID",
    "ParentLocation": "RegionName",
    "SpatialDimValueCode": "CountryID",
    "Location": "CountryName",
    "Period": "Year",
    "IndicatorCode": "IndicatorID",
    "Indicator": "IndicatorName"
}, inplace=True)

# -------------------------------------------------
# 6️⃣ Create Dimension Tables
# -------------------------------------------------

# Regions
Regions = (
    df[["RegionID", "RegionName"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

# Countries
Countries = (
    df[["CountryID", "CountryName", "RegionID"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

# Date
Date = (
    df[["Year"]]
    .drop_duplicates()
    .sort_values("Year")
    .reset_index(drop=True)
)

# Indicator
Indicator = (
    df[["IndicatorID", "IndicatorName"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

# Disaggregation (none present → default Total)
Disaggregation = pd.DataFrame({
    "DisaggregationID": [1],
    "Value": ["Total"]
})

# -------------------------------------------------
# 7️⃣ Create Fact Table (Indicator Value)
# -------------------------------------------------
IndicatorValue = df[["CountryID", "Year", "IndicatorID", "Value"]].copy()

IndicatorValue["DisaggregationID"] = 1

# Add primary key RowID
IndicatorValue.insert(0, "RowID", range(1, len(IndicatorValue) + 1))

# Ensure correct types
IndicatorValue["Year"] = pd.to_numeric(IndicatorValue["Year"], errors="coerce").astype("Int64")
IndicatorValue["Value"] = pd.to_numeric(IndicatorValue["Value"], errors="coerce")

# -------------------------------------------------
# Final Tables:
# Regions
# Countries
# Date
# Indicator
# Disaggregation
# IndicatorValue
# -------------------------------------------------
```
